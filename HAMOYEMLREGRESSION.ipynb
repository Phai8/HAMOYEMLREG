 NO. 13. THE MEAN ABSOLUTE ERROR(IN TWO DECIMAL PLACES)

#ANSWER : 0.05

```python
ED2 = ED.drop(columns = ['date', 'lights'])
ED2.head(3)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
normalised = pd.DataFrame(scaler.fit_transform(ED2), columns = ED2.columns)
features = normalised.drop(columns = ["Appliances"])
features.columns
```

```python
target = normalised['Appliances']
target.head(3)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size = 0.3, random_state = 42)
x_train, x_test, y_train, y_test
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
#fit the model to the training dataset
linear_model.fit(x_train, y_train)
#obtaim prediction
y_preds = linear_model.predict(x_test)
y_preds
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_preds).round(2)
mae
```

NO.17. WHICH FEATURES HAVE THE LOWEST AND HIGHEST WEIGHT RESPECTIVELY

#ANSWER : RH_2 and RH_1

```python
def get_weights_df(model, feat, col_name):
    
#THIS fxn returns the wts of every features
    wts = pd.Series(model.coef_, feat.columns).sort_values()
    wts_df = pd.DataFrame(wts).reset_index()
    wts_df.columns = ['Features', col_name]
    wts_df[col_name].round(3)
    return wts_df
```

```python
lin_model_wts = get_weights_df(linear_model, x_train, 'Lin_Model_wts')
lin_model_wts.sort_values(by = 'Lin_Model_wts', ascending = 'False')
```

NO.18. IS THERE ANY CHANGE TO THE ROOT MEAN SQUARED ERROR (RMSE) WHEN EVALUATED ON THE TEST SET

NO.19.TRAIN A LASSO REGRESSION MODEL WITH AN ALPHA VALUE 0.001 AND OBTAIN NEW FEATURE WEIGHTS WITH IT. HOW MANY OF THE FEATURE HAVE NON-ZERO FEATURE WEIGHTS

#ANSWER : 4

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha = 0.001)
lasso_reg.fit(x_train, y_train)

def get_weights_df(model, feat, col_name):
    
#THIS fxn returns the wts of every features
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(2)
    return weights_df
```

```python
lasso_weights_df = get_weights_df(lasso_reg, x_train, 'Lasso_Weight')
lasso_weights_df
```
